```python
import pandas as pd
```

```python
ramen_ratings = pd.read_csv("ramen_ratings.csv")
sports_ratings = pd.read_csv("sports_ratings.csv")
spotify_ratings = pd.read_csv("spotify_ratings.csv")
```

```python
ramen_ratings = ramen_ratings.drop(ramen_ratings[ramen_ratings.Stars == "Unrated"].index)
```

```python
ramen_ratings = ramen_ratings.drop(columns=["Review #","Unnamed: 6"])
```

```python
ramen_ratings["Stars"].unique()
```

```python
ramen_ratings = ramen_ratings.replace("USA","United States")
```

```python
sports_ratings = sports_ratings.rename(columns={"State(s)":"Country"})
```

```python
sports_ratings = sports_ratings.drop([10,25,30])
```

```python
sports_ratings = sports_ratings.drop(columns=["Ref","Pyramid level"])
```

```python
sports_ratings["Total attendance"] = sports_ratings["Total attendance"].str.replace(',','')
```

```python
sports_ratings = sports_ratings.replace(["1465432[135]","3887562[149]","1384820[162]","959914[164]","737782[F]"],["1465432","3887562","1384820","959914","737782"])
sports_ratings = sports_ratings.replace("295[140]","295")
```

```python
sports_ratings["Year"] = sports_ratings["Season"].str.split("–").str[0]
```

```python
sports_ratings = sports_ratings.replace("2018-2019","2018")
```

```python
sports_ratings = sports_ratings.drop(columns=["Season"])
```

```python
sports_ratings = sports_ratings.astype({"Teams":int,"Games":int,"Total attendance":int,"Average attendance":int,"Year":int})
```

```python
sports_ratings["Country"].unique()
```

```python
spotify_ratings
```

```python
spotify_ratings = spotify_ratings.rename(columns={"country":"Country"})
```

```python
spotify_ratings["streams"] = spotify_ratings["streams"].str.replace(',','')
```

```python
spotify_ratings = spotify_ratings.replace("USA","United States")
```

```python
spotify_ratings = spotify_ratings.astype({"streams":int,"rank":int})
```

```python
spotify_ratings.to_csv("data/spotify_ratings_final.csv",index=False)
```
