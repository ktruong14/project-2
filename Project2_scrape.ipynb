```python
# Dependencies
import pandas as pd
```

```python
# URLs of page to be scraped
url_netflix = 'https://www.comparitech.com/tv-streaming/netflix-subscribers/'
URL_sports = "https://en.wikipedia.org/wiki/List_of_attendance_figures_at_domestic_professional_sports_leagues"
```

# Scrape Netflix Subscriber Numbers by Country (q1 2020)

```python
#scrape the netflix subscribers data table
netflix_tables = pd.read_html(url_netflix)
print("There are : ",len(netflix_tables)," tables")
netflix_tables[1]
```

```python
# make netflix table into df
netflix_df = netflix_tables[1]
```
